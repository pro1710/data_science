```python
from sklearn import datasets
%pylab inline
```

```python
circles = datasets.make_circles()
```

```python
print('features: \n{}\ntarget: \n{}'.format(circles[0][:10], circles[1][:10]))
```

```python
from matplotlib.colors import ListedColormap
```

```python
def plot_2d_dataset(data, colors):
    pylab.figure(figsize(8, 8))
    x = list(map(lambda x: x[0], data[0]))
    y = list(map(lambda x: x[1], data[0]))
    pylab.scatter(x, y, c=data[1], cmap=colors)
```

```python
colors2 = ListedColormap(['red', 'yellow'])
plot_2d_dataset(circles, colors2)
```

```python
noisy_circles = datasets.make_circles(noise=0.15)
plot_2d_dataset(noisy_circles, colors2)
```

```python
simple_classification_problem = datasets.make_classification(n_features=2, 
                                                             n_informative=1,
                                                            n_redundant=1,
                                                            n_clusters_per_class=1,
                                                            random_state=1)
```

```python
plot_2d_dataset(simple_classification_problem, colors2)
```

```python
classification_problem = datasets.make_classification(n_features=2,
                                                     n_informative=2,
                                                     n_classes=4,
                                                     n_redundant=0,
                                                     n_clusters_per_class=1, 
                                                     random_state=1)
```

```python
colors4 = ListedColormap(['red', 'blue', 'green', 'yellow'])
```

```python
plot_2d_dataset(classification_problem, colors4)
```
